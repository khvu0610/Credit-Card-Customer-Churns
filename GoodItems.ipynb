```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
# read data
df = pd.read_csv('../Data/Gooditem.csv')
df.head()
```

```python
df.info()
```

```python
# vẽ top 10 sản phẩm bán chạy nhất
df['Product Name'].value_counts().head(10).plot.barh(figsize=(10, 5))
plt.title('Top 10 sản phẩm bán chạy nhất', fontsize=20)
plt.xlabel('Tên sản phẩm', fontsize=15)
plt.ylabel('Số lượng', fontsize=15)
plt.gca().invert_yaxis()
plt.show()
```

```python
# vẽ biểu đồ tần suất sử dụng kho
df['Warehouse_block'].value_counts().plot.barh(figsize=(10, 5))
plt.title('Biểu đồ tần suất sử dụng kho', fontsize=20)
plt.ylabel('Tên kho', fontsize=15)
plt.xlabel('Số lượng', fontsize=15)
plt.gca().invert_yaxis()
plt.show()
```

```python
# biểu đồ thể hiện chênh lệch doanh thu và chi phí vận chuyển
df['Profit'] = df['Cost_of_the_Product'] - df['Product Price']
# vẽ biểu đồ
plt.figure(figsize=(10, 10))
sns.barplot(y='Product Name', x='Profit', data=df)
plt.title('Chênh lệch doanh thu và chi phí vận chuyển', fontsize=20)
plt.ylabel('Tên sản phẩm', fontsize=15)
plt.xlabel('Chênh lệch', fontsize=15)
# plt.xticks(rotation=90)
plt.show()
```
