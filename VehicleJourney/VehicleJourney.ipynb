```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
# read data
df = pd.read_csv('../Data/VehicleJourney.csv')
df.head()
```

```python
df.info()
```

```python
# biểu đồ pie về tình trong delay
df['delay'].value_counts().plot(kind='pie', figsize=(10, 10), autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Delay')
plt.show()
```

Nhận xét: tình trạng giao hàng bị trễ thường xuyên xảy ra (lên đến 71,9 % trên toàn đơn hàng)

```python
# biểu đồ các mặt hàng thường bị delay
df[df['delay'] == 'Yes']['Product Name'].value_counts().plot(kind='barh', figsize=(10, 10))
plt.title('Các mặt hàng thường bị delay')
plt.xlabel('Số lượng')
plt.ylabel('Tên mặt hàng')
plt.gca().invert_yaxis()
plt.show()
```
